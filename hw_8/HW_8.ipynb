Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import pandas as pd
```

```python
# Нормальное распределение. изменяем ср.значение(мю)
mu = range(3,18,3)
sigma = 1
x = np.arange(0,20,.1)
fig = plt.figure(figsize=(12,5))

for mu_i in mu:
    normal_disribution = sts.norm(loc = mu_i, scale = sigma)
    normal_disribution_pdf = normal_disribution.pdf(x)
    plt.plot(x,normal_disribution_pdf)

plt.legend(list(map(lambda x: f'mu = {x}',list(mu))))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal Distibution. \n varing mu');
```

```python
# Нормальное распределение. изменяем Отклонение(сигму)
mu = 10
sigma = range(1,6,1)
x = np.arange(0,20,.1)
fig = plt.figure(figsize=(12,5))

for sigma_i in sigma:
    normal_disribution = sts.norm(loc = mu, scale = sigma_i)
    normal_disribution_pdf = normal_disribution.pdf(x)
    plt.plot(x,normal_disribution_pdf)

plt.legend(list(map(lambda x: f'sigma = {x}',list(sigma))))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Normal Distibution. \n varing sigma');
```

### Экспоненциальное распределение

```python
# Экспоненциальное распределение. изменяем  частоту(лямбда)
lambda_ = range(1,18,3)
x = np.arange(0,20,.1)
fig = plt.figure(figsize=(12,5))

for lambda_i in lambda_:
    expon_disribution = sts.expon(lambda_i)
    expon_disribution_pdf = expon_disribution.pdf(x)
    plt.plot(x,expon_disribution_pdf)

plt.legend(list(map(lambda x: f'lambda = {x}',list(lambda_))))
plt.xlabel('x')
plt.xticks(range(0,21,1))
plt.ylabel('f(x)')
plt.title('Exponential Distibution. \n varing lambda - frequency');
```

```python
# Экспоненциальное распределение. изменяем  Отклонение(сигму)
lambda_ = 1 
scale_ = range(1,18,3)
x = np.arange(0,10,.1)
fig = plt.figure(figsize=(12,5))

for scale_i in scale_:
    expon_disribution = sts.expon(lambda_, scale=scale_i)
    expon_disribution_pdf = expon_disribution.pdf(x)
    plt.plot(x,expon_disribution_pdf)

plt.legend(list(map(lambda x: f'scale = {x}',list(scale_))))
plt.xlabel('x')
plt.xticks(np.arange(0,11,.5))
plt.ylabel('f(x)')
plt.title('Exponential Distibution. \n varing deviation');
```

### Распределение Стьюдента

```python
# Распределение Стьюдента. изменяем количество испытаний в выборке(n)
n = range(3,35,7)  #Количество испытаний в выборке
mu = 0             #Среднее значение
sigma = 1          #Отклонение 
x = np.linspace(-5,5,100)
fig = plt.figure(figsize=(12,5))

for n_i in n:
    t_disribution = sts.t(df= n_i, loc=mu, scale = sigma)
    t_disribution_pdf = t_disribution.pdf(x)
    plt.plot(x,t_disribution_pdf)

plt.legend(list(map(lambda x: f'n = {x}',list(n))))
plt.xlabel('x')
plt.xticks(np.arange(-5,5.1,.5))
plt.ylabel('f(x)')
plt.title('Student Distibution. \n varing number of samples(n)');
```

```python
# Распределение Стьюдента. изменяем среднее(mu)
n = 10             #Количество испытаний в выборке
mu = range(0,12,3) #Среднее значение
sigma = 1          #Отклонение 
x = np.linspace(-5,15,100)
fig = plt.figure(figsize=(12,5))

for mu_i in mu:
    t_disribution = sts.t(df= n, loc=mu_i, scale = sigma)
    t_disribution_pdf = t_disribution.pdf(x)
    plt.plot(x,t_disribution_pdf)

plt.legend(list(map(lambda x: f'mu = {x}',list(mu))))
plt.xlabel('x')
plt.xticks(np.arange(-5,15.1,1))
plt.ylabel('f(x)')
plt.title('Student Distibution. \n varing central tendency(mu)');
```

```python
# Распределение Стьюдента. изменяем Отклонение(sigma)
n = 10                #Количество испытаний в выборке
mu = 0                #Среднее значение
sigma = range(0,6,1) #Отклонение 
x = np.linspace(-10,10,100)
fig = plt.figure(figsize=(12,5))

for sigma_i in sigma:
    t_disribution = sts.t(df= n, loc=mu, scale = sigma_i)
    t_disribution_pdf = t_disribution.pdf(x)
    plt.plot(x,t_disribution_pdf)

plt.legend(list(map(lambda x: f'mu = {x}',list(sigma))))
plt.xlabel('x')
plt.xticks(np.arange(-10,10.1,1))
plt.ylabel('f(x)')
plt.title('Student Distibution. \n varing deviation(sigma)');
```

- Задание 2

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания  в отрезок $[3.5;7]$, т.е. P(3.5 &lt; $\xi$ &lt; 7).

#### Решение: 

$$ C = \frac{1}{5-0} = \frac{1}{5} $$

Так как интеграл плотностей от минус бесконечности до плюс бесконечности должен равняться 1 (условие функции распределения) вероятность не может быть больше 1 (или 100%):
$$ F(x) = \int_{-\infty}^{\infty} f(\xi) d\xi = 1$$

Учитывая, что интеграл - это площадь под кривой, а кривая представляет собой прямоугольник, одна из сторон которого 
$$ b = 5-0 = 5. $$  

То находим другую сторону прямоугольника, как  
$$ S = b \cdot C \\
\frac{площадь}{сторону} = \frac{S}{b} = \frac{1}{5} $$

Стало быть
$$ C = \frac{1}{5-0} = \frac{1}{5} $$
Математическое ожидание: 
$$ M(x) = \int_{-\infty}^{\infty}\xi \cdot f(\xi) d\xi =  \frac{1}{5} \cdot \frac{1}{2} (\xi^2) \bigg|_0^5 = \frac{1}{5} \cdot \frac{1}{2} \cdot (25-0) = \frac{25}{10} = 2.5$$

Плотность попадания в отрезок $[3.5;7]$ :
$$ P(3.5 < \xi < 7) = \int_{3.5}^{7}f(\xi) d\xi = \frac{1}{5} \cdot \xi \bigg|_{3.5}^{5} + 0= \frac{1}{5} \cdot  \frac{15}{10} = 0.3$$

Проверим в ноутбуке:

```python
uniform_distribution = sts.uniform(0,5)
xi = np.arange(-1,8,0.1)
uniform_distribution_pdf = uniform_distribution.cdf(xi)
#график
plt.plot(xi, uniform_distribution_pdf)
#границы
plt.plot([3.5,3.5],[0,1.1])
plt.plot([7,7],[0,1.1])
#обозначения
plt.yticks(np.arange(0,1.1,.1))
plt.xlim([0,7.5])
plt.ylim([0,1.05])
plt.xlabel('xi')
plt.ylabel('F(xi)')

print( f' Плотность попадания в отрезок [3.5;7] = {(uniform_distribution.cdf(7) - uniform_distribution.cdf(3.5)):.1f} ')
```

```python
mean, var, _, _ = uniform_distribution.stats(moments='mvsk')
print(f'Мат ожидание M(x) = {mean}, отклонение (sigma) = {var**0.5 :.2f}')
```

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
#первая задача:
N = 54 + 4 + 28
p1 = 54/N
p2 = 4/N
p3 = 28/N
I = -p1 * np.log2(p1) - p2 * np.log2(p2) - p3 * np.log2(p3)
print('Первая задача в случае 54, 4 и 28 секунд.')
print( f'Сообщение о цвете светофора в текущий момент содержит {I :.2f} бит информации.')
```

```python
#вторая задача:
N = 3
p = 1/3

I = -p * np.log2(p)

print('Вторая задача (в случае 28, 28 и 28 секунд).')
print( f'Сообщение о цвете светофора в текущий момент содержит {I :.2f} бит информации.')
```
