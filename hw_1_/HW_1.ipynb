# Домашнее задание "Линейная алгебра. Векторы".

Даны вектора x = (1,1), y = (2, 0) и z = (0,2)

Надо:
- Изобразить вектора на экране
- Изобразить точку x + y + z
- Найти угол между векторами x и y



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])
```

### Изобразить вектора на экране

```python
ax = plt.axes()

ax.arrow(0,0,x[0],x[1],head_width = 0.1)
ax.annotate("x", xy=(x[0],x[1]), xytext=(x[0]-.5,x[1]+.5), 
            arrowprops=dict(arrowstyle="->") )
ax.arrow(0,0,y[0],y[1],head_width = 0.1)
ax.annotate("y", xy=(y[0],y[1]), xytext=(y[0]-.5,y[1]+.5), 
            arrowprops=dict(arrowstyle="->") )
ax.arrow(0,0,z[0],z[1],head_width = 0.1)
ax.annotate("z", xy=(z[0],z[1]), xytext=(z[0]-.5,z[1]+.5), 
            arrowprops=dict(arrowstyle="->") )

plt.grid()
```

### Изобразить точку x + y + z

```python
c = x+y+z
```

```python
plt.plot(c[0], c[1], 'o')
```

```python
ax = plt.axes()

ax.arrow(0,0,x[0],x[1],head_width = 0.1)
ax.annotate("x", xy=(x[0],x[1]), xytext=(x[0]-.5,x[1]+.5), 
            arrowprops=dict(arrowstyle="->") )
ax.arrow(0,0,y[0],y[1],head_width = 0.1)
ax.annotate("y", xy=(y[0],y[1]), xytext=(y[0]-.5,y[1]+.5), 
            arrowprops=dict(arrowstyle="->") )
ax.arrow(0,0,z[0],z[1],head_width = 0.1)
ax.annotate("z", xy=(z[0],z[1]), xytext=(z[0]-.5,z[1]+.5), 
            arrowprops=dict(arrowstyle="->") )
#x+y+z
ax.arrow(0,0,c[0],c[1],head_width = 0.1)
ax.annotate("x+y+z", xy=(c[0],c[1]), xytext=(2,3), 
            arrowprops=dict(arrowstyle="->") )
plt.grid()
```

### Найти угол между векторами x и y

```python
norm_x = np.sqrt( (x*x).sum())
norm_y = np.sqrt( (y*y).sum()) 
summ_x_y = np.dot(x,y)
cos_alhpa = summ_x_y / (norm_x * norm_y) 
alpha_rad = np.arccos(cos_alhpa)
alpha = np.rad2deg(alpha_rad)
print('Угол между векторами x и y = %.1f'% alpha)
```
