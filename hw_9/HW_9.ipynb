```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.norm(loc = 1, scale = 1) 
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

### нормальное распределение

```python
# Параметры запуска: размеры выборок и количество повторов генерации

n = range(2,51,2)

rv = sts.norm(loc = 1, scale = 1)

samples_count = 1000
# Массивы для сохранения средних по выборкам
a = np.array([])

for n_i in n:
    a1 = np.array([])
    
    for number in range(samples_count):
        # На каждом цикле генерируем выборки разных размеров
        sample = rv.rvs(size=n_i)
        
        # Считаем среднее по каждой выборке и добавляем в массив средних
        a1 = np.append(a1, sample.mean())
        
    a = np.append(a, a1)

a = a.reshape(len(n),samples_count)
```

```python
# Строим графики
# Подписи для графиков
fig = plt.figure(figsize=(14,3))

for i,n_i in enumerate(n):
    
    fig.add_subplot(131 + i%3 )
    
    label_i = 'sample of ' + str(n_i) + ' values. '
    #plt.hist(a[i-1,:], density=True,bins=30, label=label_i)
    sns.histplot(a[i],kde=True)
    plt.title(label_i)
    plt.ylabel('samples means')
    plt.xlabel('$x$')
    
    if (i+1)%3 == 0: 
        plt.show()
        fig = plt.figure(figsize=(14,3))
    
    

```
